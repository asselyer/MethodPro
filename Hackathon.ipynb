```python
import numpy as np
import pandas as pd
import string
```

```python
data = pd.read_csv('./data/sentiment-news-bahasa-v3.csv')
```

```python
mapping = {'Positive':1, 'Neutral': 0, 'Negative': -1, 'negative':1}
data["label"] = data["label"].map(lambda x: mapping[x])
```

```python
data
```

```python
def parseOutText(content):
    words = ""
    if len(content) > 1:
        ### remove punctuation
        text_string = content.translate(str.maketrans('', '', string.punctuation))


        ### project part 2: comment out the line below
    #         words = text_string

        ### split the text string into individual words, stem each word,
        ### and append the stemmed word to words (make sure there's a single
        ### space between each stemmed word)

#         stemmer = SnowballStemmer("english")
#         text_string = content;

        words = "";
        splitted_text = text_string.split()
        for word in splitted_text:
#             word = stemmer.stem(word)
            words = words + word + " "
    
    return words


    
    
```

```python
import string
word_data = []
print(parseOutText("Ah!!! I like the weather!!!!, &&"))
for text in data["text"]:
    word_data.append(parseOutText(text) )
print(word_data)
```

```python
f = open('./data/stopwords-ms.txt', 'r')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, TfidfTransformer

vectorizer = TfidfVectorizer(f.read())
X = vectorizer.fit_transform(word_data)
transformer = TfidfTransformer()
tfidf = transformer.fit_transform(X)

words = vectorizer.get_feature_names()
print(tfidf)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score, precision_score, accuracy_score
x_train, x_test, y_train, y_test = train_test_split(tfidf, data["label"], test_size = 0.2, random_state = 7)
clf = SVC()
clf.fit(x_train, y_train)
pred = clf.predict(x_test)

print(precision_score(pred, y_test, average = 'weighted') )
print(recall_score(pred, y_test, average = 'weighted') )
print(accuracy_score(pred, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
x_train, x_test, y_train, y_test = train_test_split(tfidf, data["label"], test_size = 0.2, random_state = 7)
clf_forest = RandomForestClassifier(n_estimators = 100, random_state = 7)
clf_forest.fit(x_train, y_train)
pred = clf_forest.predict(x_test)

print(precision_score(pred, y_test, average = 'weighted') )
print(recall_score(pred, y_test, average = 'weighted') )
print(accuracy_score(pred, y_test))
```
